```python
import numpy as np
import skimage as ski
import matplotlib.pyplot as plt
```

```python
# Utils

# Importer une image
def import_image(filename):
    image = ski.io.imread(filename)
    return np.array(image)

# Afficher une image
def display_image(image, title):
    plt.figure()
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Calculer le PSNR entre deux images
def psnr(original_image, processed_image):
    original_image = original_image.astype(np.float64)
    processed_image = processed_image.astype(np.float64)
    mse = np.mean((original_image - processed_image) ** 2)
    if mse == 0:
        return float('inf')
    max_pixel = 255.0
    psnr_value = 10 * np.log10((max_pixel ** 2) / mse)
    return psnr_value
```

```python
image = import_image("lena.pgm")
display_image(image, "Image originale")

# Bruit
image_noised = ski.util.random_noise(image, mean=0, var=0.01) * 255
image_noised = image_noised.astype(np.uint8)
display_image(image_noised, 'Image bruitée')
filtered_image = image_noised.copy()

# Définir la taille de la fenêtre (3x3 dans cet exemple)
window_size = 3
offset = window_size // 2

# Parcourir chaque pixel de l'image
for i in range(offset, image_noised.shape[0] - offset):
    for j in range(offset, image_noised.shape[1] - offset):
        # Fenêtre autour du pixel
        window = image_noised[i - offset:i + offset + 1, j - offset:j + offset + 1]
        
        # Médiane
        median_value = np.median(window)
        
        filtered_image[i, j] = median_value

display_image(filtered_image, "Image après filtre médian")

plt.show()
```

```python
image = import_image("lena.pgm")
display_image(image, "Image originale")

# Bruit
image_noised = ski.util.random_noise(image, mean=0, var=0.01) * 255
image_noised = image_noised.astype(np.uint8)
display_image(image_noised, 'Image bruitée')
filtered_image = image_noised.copy()

# Définir la taille de la fenêtre (3x3 dans cet exemple)
window_size = 3
pad_size = window_size // 2

# Ajouter un padding à l'image pour traiter les bords
padded_image = np.pad(image, pad_size, mode='edge')

# Parcourir chaque pixel de l'image
for i in range(pad_size, padded_image.shape[0] - pad_size):
    for j in range(pad_size, padded_image.shape[1] - pad_size):
        # Extraire la fenêtre autour du pixel (i, j)
        window = padded_image[i - pad_size:i + pad_size + 1, j - pad_size:j + pad_size + 1]
        
        # Calculer la médiane de la fenêtre
        median_value = np.median(window)
        
        # Assigner la médiane au pixel correspondant dans l'image filtrée
        filtered_image[i - pad_size, j - pad_size] = median_value

display_image(filtered_image, "Image après filtre médian")

plt.show()
```
